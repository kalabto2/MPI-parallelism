```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
col_names = ["file_path", "n", 'e', "solution", "num_solutions", "time", "num_thread", "num_proc"]
```

```python
def print_graph(df, title):
    fig = plt.figure(figsize=(10, 6))
    
    i = 0
    for f in df["file_path"].unique():
#         if i == 1:
#             i += 1
#             continue
            
        print(f)
        i += 1
        ff = df[df["file_path"] == f]
        
        fp = ff["file_path"].to_list()[0]
        times = ff['time'].to_list()
        threads = ff['num_thread'].to_list()
        print(times)
        print(threads)
        plt.plot(threads, times, label=fp)
    
    plt.legend()
    plt.xlabel("počet vláken")
    plt.ylabel("čas (s)")
    fig.suptitle(title, fontsize=16)
    fig.tight_layout()
    plt.show()
    
    image_name = title.lower().replace(" ", "_")
    fig.savefig(f"report/images/{image_name}.png", dpi=400)
    
def print_latex(df):
    new_df = df[["file_path", "time", "num_thread", "num_proc"]]
    display(new_df)
    print(new_df.to_latex(index=False))
```

### Serial

```python
sf = pd.read_csv("data/serial.csv", sep=";", names=col_names)
display(sf)
print_latex(sf)
```

### Task

```python
tf = pd.read_csv("data/task1.csv", sep=";", names=col_names)
print_latex(tf)
print_graph(tf, "Porovnání task paralelizace")
```

### Data

```python
data_df = pd.read_csv("data/data1.csv", sep=";", names=col_names)
display(data_df)
print_latex(data_df)
print_graph(data_df, "Porovnání data paralelizace")
```

### Process

```python
pf = pd.read_csv("data/process1.csv", sep=";", names=col_names)
display(pf)
print_latex(pf)
print_graph(pf, "Porovnání process paralelizace")
```
